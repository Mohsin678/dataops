```python

#define cat and num columns

numeric_features = [feature for feature in df.columns if df[feature].dtype !="O"]
cat_features = [feature for feature in df.columns if df[feature].dtype =="O"]

print("we have {} numeric features {}".format(len(numeric_features),numeric_features))
print("we have {}categorical feature {}".format(len(cat_features),cat_features))

```

```python
#adding column for tot score and avg

df["tot_score"] = df["math_score"] +df["reading_score"] +df["writing_score"]
df["average"] = df["tot_score"]/3
df.head()
```

```python
reading_full =  df[df["reading_score"]==100]["average"].count()
writing_full  = df[df["writing_score"]==100]["average"].count()
math_full  = df[df["math_score"]==100]["average"].count()

print(f"numb of students full marks in reading{reading_full}")
print(f"numb of students full marks in writing{writing_full}")
print(f"numb of students full marks in math{math_full}")
```

```python
read_20 = df[df["reading_score"]<=20]["average"].count()
write_20 = df[df["writing_score"]<=20]["average"].count()
math_20 = df[df["math_score"]<=20]["average"].count()
print(f"numb of student got less than 20 in reading{read_20}")
print(f"numb of student got less than 20 in writing{write_20}")
print(f"numb of student got less than 20 in math{math_20}")

```
